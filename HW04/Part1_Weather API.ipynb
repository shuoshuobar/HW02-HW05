```python
url3 = "http://api.weatherapi.com/v1/forecast.json?key=d6c0a6e1e2c34487bf405758240511&q=Heathrow&days=3&aqi=no&alerts=no"
response3 = requests.get(url3)
data3 = response3.json()
data3
```

```python
# Question 7: Print the date of each of the 3 days you're getting a forecast for.
days = [data['date'] for data in data3['forecast']['forecastday']]
print(f"The days are: {', '.join(days)}.")
```

```python
# Question 7: Print the date of each of the 3 days you're getting a forecast for.
for data in data3['forecast']['forecastday']: 
    date = data['date']
    print(f"The date is: {date}.")
```

```python
# Question 8: Print the maximum temperature of each of the days.
for data in data3['forecast']['forecastday']: # data3['forecast']['forecastday'] is the list 
    date = data['date']
    max_temp = data['day']['maxtemp_c']
    print(f"The maximum temperature of {date} is: {max_temp} celsius.")
```

```python
# Question 9: Print only the day with the highest maximum temperature.

# Assume the first day has the highest temperature
warmest_day = data3['forecast']['forecastday'][0] # warmest_day is the first item of the list
highest_temp = warmest_day['day']['maxtemp_c']

# Iterate over the remaining forecast days
for data in data3['forecast']['forecastday'][1:]: # data is an item of the list 
    date = data['date']
    max_temp = data['day']['maxtemp_c']
    
# Compare each day's max temperature to the current highest
if max_temp > highest_temp:
    highest_temp = max_temp
    warmest_day = data # warmest_day is updated to another item, data, of the list

# Print the date with the highest temperature
print(f"The day with the highest maximum temperature is {warmest_day['date']}.")
```
